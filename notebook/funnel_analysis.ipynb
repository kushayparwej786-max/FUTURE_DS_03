```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# dataset load (IMPORTANT: no sep)
df = pd.read_csv("bank.csv")

# check columns
print(df.columns)

# preview
df.head()

# Chart 1 - Customer Conversion
plt.figure(figsize=(6,4))
sns.countplot(x="deposit", data=df)
plt.title("Customer Conversion (Yes vs No)")
plt.show()

# Chart 2 - Age Distribution
plt.figure(figsize=(6,4))
sns.histplot(df["age"], bins=20)
plt.title("Age Distribution")
plt.show()

# Chart 3 - Job vs Conversion
plt.figure(figsize=(10,5))
sns.countplot(x="job", hue="deposit", data=df)
plt.xticks(rotation=90)
plt.title("Job vs Conversion")
plt.show()

# Chart 4 - Balance vs Conversion
plt.figure(figsize=(6,4))
sns.boxplot(x="deposit", y="balance", data=df)
plt.title("Balance vs Conversion")
plt.show()
```
